```python
from reader_utils import detect_img, load_VehicleNet
import os
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
train_imgs_dir = '../../../cityscapes_data/leftImg8bit/my_train/'
val_imgs_dir = '../../../cityscapes_data/leftImg8bit/my_val/'

train_imgs_files = os.listdir(train_imgs_dir)
val_imgs_files = os.listdir(val_imgs_dir)
```

```python
train_imgs_files[0]
```

```python
im = cv2.imread(os.path.join(train_imgs_dir, train_imgs_files[10]))
segmented_im = detect_img(im, load_VehicleNet())
plt.imshow(segmented_im, interpolation='nearest')
plt.show()
```

```python
model = load_VehicleNet()
N = 10
pick_imgs = np.random.choice(val_imgs_files, N)
f, axarr = plt.subplots(N, 2, figsize=(24, 24))
for i in range(N):
    im = cv2.imread(os.path.join(val_imgs_dir, pick_imgs[i]))
    segmented_im = detect_img(im, model)
    axarr[i, 0].imshow(im, interpolation='nearest')
    axarr[i, 1].imshow(segmented_im, interpolation='nearest')
```
